```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Provided data
data = {
    'Name': ['Mauro Icardi', 'Edin Dzeko', 'Paul Onuachu', 'Dal Varesanovic', 'Boa Morte', 'Sebastian Szymanski', 'Jackson Muleka', 'Omar Colley'],
    'Shoots per 90 minutes': [3.67, 5.18, 3.89, 2.25, 2.32, 2.59, 3.12, 0.7],
    'Goals per 90 minutes': [0.52, 0.89, 0.69, 0.27, 0.32, 0.56, 0.21, 0.45],
    'Average xG per game': [0.57, 0.83, 0.51, 0.4, 0.36, 0.33, 0.26, 0.3]
}

# Creating a DataFrame
df_players = pd.DataFrame(data)

# Plotting
plt.figure(figsize=(10, 6))

# Scatter plot for the data
# The size of each point is proportional to the Average xG per shoot
scatter = plt.scatter(df_players['Shoots per 90 minutes'], df_players['Goals per 90 minutes'], 
                      s=df_players['Average xG per game']*100,  # Multiply by 100 to scale up the point sizes
                      alpha=0.5, color='red')

for i in range(len(df_players)):
    plt.annotate(df_players['Name'][i],(df_players['Shoots per 90 minutes'][i],df_players['Goals per 90 minutes'][i]),textcoords="offset points", 
                 xytext=(4,4), 
                 ha='center',
                va='bottom')
    
plt.title('Player Shooting Efficiency')
plt.xlabel('Shots per 90 minutes')
plt.ylabel('Goals per 90 minutes')

# Show grid
plt.grid(True)

# Show the plot
plt.show()    
```
